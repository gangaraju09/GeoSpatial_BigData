## Loading the US states shapefile and get its geometries

```python
# Download the GIS layer from US Census and unzip files (cb_2018_us_state_20m.zip) to your local directories
# https://www.census.gov/geographies/mapping-files/time-series/geo/carto-boundary-file.html
import geopandas as gpd # import geopandas
USA = gpd.read_file("./shapefiles/cb_2018_us_state_20m.shp") # read data
USA
```

```python
USA.plot()
```

```python
# print the Hawaii state 
hawaii = USA[USA['NAME'].str.contains('Hawaii')]
hawaii
```

```python
hawaii.plot()
```

```python
# print the Hawaii state geometry using WKT format
import pandas as pd
pd.options.display.max_colwidth=50
print(pd.options.display.max_colwidth) #check the default display width for the dataframe 
```

```python
# print the Hawaii state geometry using WKT format
print(hawaii['geometry'].to_wkt())

pd.options.display.max_colwidth = None ## display wider string

with open ('hawaii.txt', 'w') as fw:
    fw.write(str(hawaii['geometry'].to_wkt())) #output to a local file
```
