```python
import geopandas as gpd # import geopandas
rivers = gpd.read_file("./shapefiles/Rivers.shp") # read data
rivers
```

```python
rivers.plot()
```

## Count the total number of points/vertices across all 23 rivers in this map layer

```python
from shapely.geometry import mapping

total_points = 0 

for index, row in rivers.iterrows():
    shape  = row['geometry']
    print(mapping(shape)) # get JSON format
    #each Polyline/Polygon shape is a single ordered list of points#
    points = mapping(shape)['coordinates']
    total_points += len(points)
print("total points:", total_points)

```

## Compute the total lengths of all 23 rivers in this map layer

```python
total_length = 0 

for index, row in rivers.iterrows():
    shape  = row['geometry']
    #each Polyline/Polygon shape is a single ordered list of points#
    length = shape.length
    print(index,length)
    total_length += length
print("total length:", total_length)
```
