```python
import networkx as nx
import numpy as np
import pandas as pd
import osmnx as ox
import geopandas as gpd
import requests
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import matplotlib.colors as colors
ox.config(use_cache=True, log_console=True)
ox.__version__
```

```python
# get the administrative boundaries for a place
campus = ox.geocode_to_gdf('UW-Madison, WI, USA')
ax = ox.project_gdf(campus).plot()
_ = ax.axis('off')
campus.to_file('uwmadison.shp')
```

```python
# get the administrative boundaries for a place
campus = ox.geocode_to_gdf('UW-Madison, WI, USA')
ax = ox.project_gdf(campus).plot()
_ = ax.axis('off')
campus.to_file('uwmadison.shp')
```

```python
# get the network for a place
campus_G = ox.graph_from_place('UW-Madison, WI, USA')
fig, ax = ox.plot_graph(campus_G)
ox.save_graph_shapefile(campus_G, 'uwmadison')
```

```python
uwmadison = gpd.read_file('uwmadison.shp')
uwmadison.plot()
```

```python
uwmadison_edge = gpd.read_file('uwmadison/edges.shp')
uwmadison_edge.plot()
```

```python
uwmadison_node = gpd.read_file('uwmadison/nodes.shp')
uwmadison_node.plot(markersize=0.1)
```

```python
# convert graph to line graph so edges become nodes and vice versa
campus_line_G = nx.line_graph(campus_G)
```

```python
# calculate the closeness centrality using networkx
%time
campus_edge_centrality = nx.closeness_centrality(campus_line_G)
```

```python
# calculate the basic statistics using osmnx
basic_stats_campus_G = ox.basic_stats(campus_G)
basic_stats_campus_G
```

```python
# calculate the extended statistics using osmnx, cc=True means calculate the closeness centrality
extended_stats_campus_G = ox.extended_stats(campus_G, cc=True)
extended_stats_campus_G
```

```python
print(extended_stats_campus_G.keys())
print(extended_stats_campus_G["closeness_centrality"])
```

```python
G = ox.graph_from_place('Piedmont, CA, USA', network_type='drive')
fig, ax = ox.plot_graph(G)
ox.save_graph_shapefile(G, 'Piedmont')
```

```python
# get a graph for some city
G = ox.graph_from_place('Madison, WI, USA', network_type='drive')
fig, ax = ox.plot_graph(G)
```

```python
%%time
edge_centrality = nx.closeness_centrality(nx.line_graph(G))
```

```python
# list of edge values for the orginal graph
ev = [1 for edge in G.edges()]

# color scale converted to list of colors for graph edges
norm = colors.Normalize(vmin=min(ev)*0.8, vmax=max(ev))
cmap = cm.ScalarMappable(norm=norm, cmap=cm.inferno)
ec = [cmap.to_rgba(cl) for cl in ev]

# color the edges in the original graph with closeness centralities in the line graph
fig, ax = ox.plot_graph(G, bgcolor='k', node_size=0,
                        edge_color=ec, edge_linewidth=1.5, edge_alpha=1)
```

```python
# color the edges in the original graph with closeness centralities in the line graph
fig, ax = ox.plot_graph(G, bgcolor='r', node_size=10,
                        edge_color=ec, edge_linewidth=1.5, edge_alpha=1)
```

```python
# calculate the shortest path and plot the route
route = nx.shortest_path(G, 53477376, 652738561)
ox.plot_graph_route(G, route)
```
