```python
import numpy as np
import libpysal as ps
from mgwr.gwr import GWR, MGWR
from mgwr.sel_bw import Sel_BW
import geopandas as gp
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
```

```python
#Load Georgia dataset and generate plot of Georgia counties (figure 1)
georgia_data = pd.read_csv(ps.examples.get_path('GData_utm.csv'))
georgia_shp = gp.read_file(ps.examples.get_path('G_utm.shp'))
fig, ax = plt.subplots(figsize=(10,10))
georgia_shp.plot(ax=ax, **{'edgecolor':'black', 'facecolor':'white'})
georgia_shp.centroid.plot(ax=ax, c='black')
```

```python
#Prepare Georgia dataset inputs
g_y = georgia_data['PctBach'].values.reshape((-1,1))
g_X = georgia_data[['PctFB', 'PctBlack', 'PctRural']].values
u = georgia_data['X']
v = georgia_data['Y']
g_coords = list(zip(u,v))

g_X = (g_X - g_X.mean(axis=0)) / g_X.std(axis=0)

g_y = g_y.reshape((-1,1))

g_y = (g_y - g_y.mean(axis=0)) / g_y.std(axis=0)
```

```python
#Calibrate GWR model
gwr_selector = Sel_BW(g_coords, g_y, g_X)
gwr_bw = gwr_selector.search(bw_min=2)
print(gwr_bw)
gwr_results = GWR(g_coords, g_y, g_X, gwr_bw).fit()
```

```python
## More information about the GWR output
## https://mgwr.readthedocs.io/en/latest/generated/mgwr.gwr.GWRResults.html
## How to index a two-dimensional array: https://numpy.org/doc/stable/user/basics.indexing.html
print(gwr_results.params.shape) # get the shape (# of locations, # of Coeffiicents); (number of rows, number of columns)
gwr_results.params[0:,0]
```

```python
# Get the spatially distributed lcal R-square values
gwr_results.localR2
```

```python
gwr_results.summary()
```

## Mapping the GWR results

```python
## Plot out the spatially distributed local R-square values on a map
georgia_shp['R_square'] =gwr_results.localR2
georgia_shp['GWR_constant'] =gwr_results.params[0:,0]
georgia_shp['GWR_PctRural'] =gwr_results.params[0:,3]

georgia_shp.head()
```

```python
# more color choices at: https://matplotlib.org/stable/tutorials/colors/colormaps.html

import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 3) # create one column, multiple row of sub-figures

georgia_shp.plot(ax=ax[0], column='R_square', cmap='plasma', legend=True, legend_kwds={'label': "R-square",
                        'orientation': "horizontal"})
georgia_shp.plot(ax=ax[1],  column='GWR_constant', cmap='viridis', legend=True, legend_kwds={'label': "GWR_constant",
                        'orientation': "horizontal"})
georgia_shp.plot(ax=ax[2],  column='GWR_PctRural', cmap='Oranges', legend=True, legend_kwds={'label': "GWR_PctRural",
                        'orientation': "horizontal"})
plt.show();
```
